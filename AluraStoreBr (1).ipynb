```python
import pandas as pd

# URLs dos arquivos
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

# Carregando os arquivos
loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Adicionando coluna identificadora
loja1["loja"] = "Loja 1"
loja2["loja"] = "Loja 2"
loja3["loja"] = "Loja 3"
loja4["loja"] = "Loja 4"

# Unindo todos os DataFrames
df = pd.concat([loja1, loja2, loja3, loja4])

# Explorando
print(df.head())
print(df.info())
print(df.describe())
```

```python
df.columns = df.columns.str.lower().str.strip()  # tudo minúsculo e sem espaços
print(df.columns)
```

```python
import pandas as pd

# Carregando os dados de cada loja
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Adicionando coluna 'loja' para cada DataFrame
loja1['loja'] = 'Loja 1'
loja2['loja'] = 'Loja 2'
loja3['loja'] = 'Loja 3'
loja4['loja'] = 'Loja 4'

# Concatenando os dados em um único DataFrame
df = pd.concat([loja1, loja2, loja3, loja4])
```

```python
menos_vendidos = df.groupby(['loja', 'Produto'])['Quantidade de parcelas'].sum().reset_index()
menos_vendidos = menos_vendidos.sort_values(['loja', 'Quantidade de parcelas'], ascending=[True, True])
print(menos_vendidos.groupby('loja').first())
```

5. Frete Médio por Loja

```python
frete_medio = df.groupby('loja')['frete'].mean()
print(frete_medio)
```

6. Vendas por Categoria

```python
categorias = df.groupby(['loja', 'categoria do produto'])['quantidade de parcelas'].sum().reset_index()
print(categorias.sort_values(['loja', 'quantidade de parcelas'], ascending=[True, False]))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Faturamento por loja
faturamento.plot(kind='bar', title='Faturamento por Loja', ylabel='R$', xlabel='Loja')
plt.show()

# Avaliações médias
avaliacoes.plot(kind='pie', autopct='%1.1f%%', title='Distribuição das Avaliações Médias')
plt.ylabel('')
plt.show()

# Dispersão: Frete Médio x Faturamento
plt.figure(figsize=(8,6))
plt.scatter(frete_medio, faturamento, s=100)
for loja in faturamento.index:
    plt.text(frete_medio[loja], faturamento[loja], loja)
plt.xlabel('Frete Médio')
plt.ylabel('Faturamento')
plt.title('Frete Médio vs Faturamento por Loja')
plt.grid(True)
plt.show()
```
